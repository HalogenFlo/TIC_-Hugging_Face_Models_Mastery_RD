```python
def check_database_summary(driver):
    # Câu lệnh Cypher đếm số lượng node theo từng Nhãn (Label)
    query = "MATCH (n) RETURN labels(n) AS labels, count(*) AS count"
    
    with driver.session() as session:
        result = session.run(query)
        print("Tóm tắt dữ liệu trong Database:")
        for record in result:
            print(f"- Nhãn (Labels): {record['labels']} | Số lượng node: {record['count']}")

# Chạy hàm (đảm bảo driver của bạn đang mở)
check_database_summary(driver)
```
